```python
import json
```

```python
with open('./NewsCR.json','r') as f:
    result = json.load(f)
```

```python
result[3]['period']
```

```python
date = '1993-01-01'
date_list = []
```

```python
for i in range(len(result)):
    if(result[i]['period'] == date):
        date_list.append(result[i]['doc'])
```

```python
len(date_list)
```

```python
date_list2 = date_list[:len(date_list)//2]
date_list2_1 = date_list2[:len(date_list2)//2]
date_list2_2 = date_list2[:len(date_list2)//2]

date_list3 = date_list[len(date_list)//2:]
date_list3_1 = date_list3[:len(date_list3)//2]
date_list3_2 = date_list3[:len(date_list3)//2]
```

```python
news_doc1 = ''
for text in date_list2_1:
    news_doc1 = news_doc1 + text 
news_doc2 = ''
for text in date_list2_2:
    news_doc2 = news_doc2 + text 

news_doc3 = ''
for text in date_list3_1:
    news_doc3 = news_doc3 + text 
news_doc4 = ''
for text in date_list3_2:
    news_doc4 = news_doc4 + text 
```

```python
len(news_doc1)
```

```python
#https://financedata.github.io/posts/python_news_cloud_text.html
#형태소 분석기
#2글자 이상의 단어만 추출
#뉴스기사 5000개 이상넘어가면 에러, 나눠서 해야함 현재 4개로 나눔
from konlpy.tag import Mecab
```

```python
engine = Mecab()
nouns1 = engine.nouns(news_doc1)
nouns2 = engine.nouns(news_doc2)
nouns3 = engine.nouns(news_doc3)
nouns4 = engine.nouns(news_doc4)
```

```python

nouns1 = [n for n in nouns1 if len(n) > 1]
nouns2 = [n for n in nouns2 if len(n) > 1]
nouns3 = [n for n in nouns3 if len(n) > 1]
nouns4 = [n for n in nouns4 if len(n) > 1]

for i in range(len(nouns2)):
    nouns1.append(nouns2[i])
for i in range(len(nouns3)):
    nouns1.append(nouns3[i])
for i in range(len(nouns4)):
    nouns1.append(nouns4[i])
```

```python
from collections import Counter

count = Counter(nouns1)
tags = count.most_common(1000)
tags[:1000]
```
