```python
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib.request as ur
import random
import time
import json
import pandas as pd
```

```python
result = pd.read_json('./final_result.json')
```

```python
data = result.drop_duplicates(['id'],keep = 'first')
```

```python
data_final = data.reset_index()
```

```python
def getinfo(good_list, bad_list):
    
    info = {}
    #webdriver 객체를 만들어 준다.
    driver = webdriver.Chrome('./chromedriver2')
    # 모든 자원이 로드될때 까지 기다리게 하는 시간을 직접 지정해 준다.
    driver.implicitly_wait(5) 
    
    for i in range(len(data_final)):
        try:
            url = 'https://www.bigkinds.or.kr/news/newsDetailView.do?newsId='+data_final['id'][i]
            driver.get(url)
            driver.implicitly_wait(5) 
            #페이지의 elements를 모두 가지고 온다.
            
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')

            company = soup.select('.col-lg-3 > div >img')[0]['alt']
            title = soup.select('.newsTitleDiv > div > h3')[0].text
            genre = soup.select('.newsTitleDiv > div > p')[0].text
            doc = soup.select('.doc_desc')[0].text
            info = {
                'searchkey' : data_final['searchkey'][i],
                'url' : url,
                'period' : data_final['searchdate'][i],
                'company' : company,
                'title' : title,
                'genre,date,name' : genre,
                'doc' : doc
            }
            good_list.append(info)
        except :
            bad_list.append(i)
#             info = {
#                 'searchkey' : data_final['searchkey'][i],
#                 'url' : url,
#                 'period' : data_final['searchdate'][i]
#             }
#             bad_list.append(info)
            continue
```

```python
good_list=[]
bad_list=[]
getinfo(good_list, bad_list)
```

```python
Doc1 = []
Doc2 = []
Doc3 = []
Doc4 = []
Doc5 = []
Doc6 = []

for i in range(20000):
    Doc1.append(good_list[i])
for i in range(20000, 40000):
    Doc2.append(good_list[i])
for i in range(40000, 60000):
    Doc3.append(good_list[i])
for i in range(60000, 80000):
    Doc4.append(good_list[i])
for i in range(80000, 100000):
    Doc5.append(good_list[i])
for i in range(100000, len(good_list)):
    Doc6.append(good_list[i])
    
json.dump(Doc1, open('./News1.json', 'w+'), ensure_ascii=False)
json.dump(Doc2, open('./News2.json', 'w+'), ensure_ascii=False)
json.dump(Doc3, open('./News3.json', 'w+'), ensure_ascii=False)
json.dump(Doc4, open('./News4.json', 'w+'), ensure_ascii=False)
json.dump(Doc5, open('./News5.json', 'w+'), ensure_ascii=False)
json.dump(Doc6, open('./News6.json', 'w+'), ensure_ascii=False)
```
