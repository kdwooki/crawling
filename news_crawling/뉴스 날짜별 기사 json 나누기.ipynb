```python
import json
```

```python
with open('./NewsCR.json','r') as f:
    result = json.load(f)
```

```python
date = '2017-09-01'
date_list = []

for i in range(len(result)):
    if(result[i]['period'] == date):
        date_list.append(result[i])
        
json.dump(date_list, open('./2017_09.json', 'w+'), ensure_ascii=False)
```
