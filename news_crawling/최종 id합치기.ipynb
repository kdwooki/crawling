```python
print(a)
for i in range(len(result)):
    if(i % 5000 == 0): print(i)
    if(result[i]['searchdate'] == searchdate[a]):
        for j in range(i, len(result)):
            if(result[j]['searchdate'] == searchdate[a]):
                if(result[i]['id'] == result[j]['id']):
                    if(result[j]['searchkey'] not in result[i]['searchkey']):
                        result[i]['searchkey'].append(result[j]['searchkey'][0])
```

```python
for i in range(len(result)):
    if(result[i]['searchdate'] ==searchdate[a]):
        result[i]['searchkey'] = set(result[i]['searchkey'])
        result[i]['searchkey'] = list(result[i]['searchkey'])
json.dump(result, open('./final_result.json', 'w+'), ensure_ascii=False)
```
